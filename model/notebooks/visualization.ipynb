# Visualising processed data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
eugene = pd.read_csv('../data/processed/eugene.csv', index_col=0)
kansas = pd.read_csv('../data/processed/kansas.csv', index_col=0)
```

```python
eugene
```

```python
plt.scatter(eugene['income'], eugene['unlinked_passenger_trips'])
plt.xlabel('Income')
plt.ylabel('Unlinked Passenger Trips')
plt.show()

plt.scatter(kansas['income'], kansas['unlinked_passenger_trips'])
plt.xlabel('Income')
plt.ylabel('Unlinked Passenger Trips')
plt.show()
```

```python
corr_matrix = eugene.corr()
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

```python
df = eugene

# Scatterplot of Unlinked Passenger Trips vs. Income
sns.scatterplot(data=df, x='income', y='unlinked_passenger_trips')
plt.show()

# Pairplot of all variables
sns.pairplot(data=df)
plt.show()

# Correlation heatmap
sns.heatmap(data=df.corr(), annot=True)
plt.show()

# Barplot of Average Unlinked Passenger Trips by Year
sns.barplot(data=df, x=df.index, y='unlinked_passenger_trips')
plt.show()
```
