# Visualising processed data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
eugene = pd.read_csv('../data/processed/eugene.csv', index_col=0)
```

```python
eugene
```

```python
plt.scatter(eugene['income'], eugene['unlinked_passenger_trips'])
plt.xlabel('Income')
plt.ylabel('Unlinked Passenger Trips')
plt.show()
```

```python
corr_matrix = eugene.corr()
sns.heatmap(corr_matrix, annot=True)
plt.show()
```
